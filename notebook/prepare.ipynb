```python
import sys
sys.path.append('../')
from datasets import load_dataset
from web.model import load_model
dataset = load_dataset("taidng/UIT-ViQuAD2.0")
```

```python
train_df= dataset['train']
```

```python
train_df = train_df.to_pandas()
train_df.head()
```

```python
def replace_empty_with_plausible(row):
    if not row['answers']['text'] and not row['answers']['answer_start']:
        row['answers']['text'] = row['plausible_answers']['text']
        row['answers']['answer_start'] = row['plausible_answers']['answer_start']
    return row

train_df = train_df.apply(replace_empty_with_plausible, axis=1)
```

```python
df = train_df[["id", "question", "answers", "context"]]
df
```

```python
df['answer'] = df['answers'].apply(lambda x: x['text'])
df['answer_start'] = df['answers'].apply(lambda x: x['answer_start'])
```

```python
# Convert list to string
df['answer'] = df['answer'].apply(lambda x: ', '.join(map(str, x)) if len(x) > 0 else '')
# Convert list to string
df['answer_start'] = df['answer_start'].apply(lambda x: ', '.join(map(str, x)) if len(x) > 0 else '')
# Remove rows with empty 'answer'
df = df[df['answer'].str.strip() != '']
df.head()
```

```python
df = df.drop(columns=['answers','id'])
df.head()
```

```python
df.to_csv('../data/custom.csv', index=False)
```
